```python
#delete file xml files and images with extra classes
import os
import glob
import re

# Define the classes to be checked and the folders containing the files
classes_to_check = ["D44","D43","D50","D01","D0w0","D11","Repair","Block crack"]
xml_folder = "D:/research/rdd2022/RDD2022_all_countries/New/xmls"
image_folder = "D:/research/rdd2022/RDD2022_all_countries/New/images"

# Get a list of all the XML files in the XML folder
xml_files = glob.glob(os.path.join(xml_folder, "*.xml"))

# Loop through each XML file
for xml_file in xml_files:
    # Read the contents of the XML file
    with open(xml_file, "r") as f:
        contents = f.read()

    # Check if the XML file contains any of the classes to be checked
    if any(class_to_check in contents for class_to_check in classes_to_check):
        # Delete the XML file
        os.remove(xml_file)
        print(f"{xml_file} has been deleted.")

        # Delete the corresponding image file, if it exists
        image_file = os.path.join(image_folder, os.path.splitext(os.path.basename(xml_file))[0] + ".jpg")
        if os.path.exists(image_file):
            os.remove(image_file)
            print(f"{image_file} has been deleted.")

print("Files containing the specified classes have been deleted!")
```

```python
import xml.etree.ElementTree as ET
import os

classes = set()

# Set the path to the folder containing the XML files
xml_folder_path ="D:/research/rdd2022/RDD2022_all_countries/New/xmls"

# Iterate over all XML files in the folder
for xml_file in os.listdir(xml_folder_path):
    if xml_file.endswith(".xml"):
        # Parse the XML file
        tree = ET.parse(os.path.join(xml_folder_path, xml_file))
        root = tree.getroot()

        # Iterate over all object tags and add the class names to the set
        for obj in root.findall('object'):
            label = obj.find("name").text
            classes.add(label)

# Print the number of unique classes
print(f"Number of unique classes: {len(classes)}")
print(f"Classes: {classes}")
```

```python
#Counting
import os
import glob
import xml.etree.ElementTree as ET

input_dir = "D:/research/rdd2022/RDD2022_all_countries/New/xmls"
class_count = {}

# loop through each xml file
for xml_file in glob.glob(os.path.join(input_dir, '*.xml')):
    # parse the xml file
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    # loop through each object in the xml file
    for obj in root.findall('object'):
        label = obj.find("name").text
        
        # if the label is new, add it to the class_count dictionary
        if label not in class_count:
            class_count[label] = 0
        
        # increment the count for the label
        class_count[label] += 1

# print the count for each class
for label, count in class_count.items():
    print(f"{label}: {count}")
```

```python
#using this file to convert from xml to text


import xml.etree.ElementTree as ET
import glob
import os
import json


def xml_to_yolo_bbox(bbox, w, h):
    # xmin, ymin, xmax, ymax
    x_center = ((bbox[2] + bbox[0]) / 2) / w
    y_center = ((bbox[3] + bbox[1]) / 2) / h
    width = (bbox[2] - bbox[0]) / w
    height = (bbox[3] - bbox[1]) / h
    return [x_center, y_center, width, height]


def yolo_to_xml_bbox(bbox, w, h):
    # x_center, y_center width heigth
    w_half_len = (bbox[2] * w) / 2
    h_half_len = (bbox[3] * h) / 2
    xmin = int((bbox[0] * w) - w_half_len)
    ymin = int((bbox[1] * h) - h_half_len)
    xmax = int((bbox[0] * w) + w_half_len)
    ymax = int((bbox[1] * h) + h_half_len)
    return [xmin, ymin, xmax, ymax]


classes = []
input_dir = "D:/research/rdd2022/RDD2022_all_countries/China_MotorBike/China_MotorBike/train/annotations/new_xmls"
output_dir = "D:/research/rdd2022/RDD2022_all_countries/China_MotorBike/China_MotorBike/train/annotations/labels"
image_dir = "D:/research/rdd2022/RDD2022_all_countries/China_MotorBike/China_MotorBike/train/images"


# create the labels folder (output directory)
if not os.path.isdir(output_dir):
    os.mkdir(output_dir)

# identify all the xml files in the annotations folder (input directory)
files = glob.glob(os.path.join(input_dir, '*.xml'))
# loop through each 
for fil in files:
    basename = os.path.basename(fil)
    filename = os.path.splitext(basename)[0]
    # check if the label contains the corresponding image file
    if not os.path.exists(os.path.join(image_dir, f"{filename}.jpg")):
        print(f"{filename} image does not exist!")
        continue

    result = []

    # parse the content of the xml file
    tree = ET.parse(fil)
    root = tree.getroot()
    width = int(root.find("size").find("width").text)
    height = int(root.find("size").find("height").text)

    for obj in root.findall('object'):
        label = obj.find("name").text
        # check for new classes and append to list
        if label not in classes:
            classes.append(label)
        index = classes.index(label)
        pil_bbox = [int(x.text) for x in obj.find("bndbox")]
        yolo_bbox = xml_to_yolo_bbox(pil_bbox, width, height)
        # convert data to string
        bbox_string = " ".join([str(x) for x in yolo_bbox])
        result.append(f"{index} {bbox_string}")

    if result:
        # generate a YOLO format text file for each xml file
        with open(os.path.join(output_dir, f"{filename}.txt"), "w", encoding="utf-8") as f:
            f.write("\n".join(result))
    else:
        # create an empty file if no object annotations are found
        with open(os.path.join(output_dir, f"{filename}.txt"), "w", encoding="utf-8") as f:
            pass


```

```python
#dividing the train set into train and val
import os
import shutil
from sklearn.model_selection import train_test_split

# Set the paths to the input and output directories
input_dir = 'D:/research/rdd2022/RDD2022_all_countries/China_Motorbike/input'
output_dir = 'D:/research/rdd2022/RDD2022_all_countries/China_Motorbike/train'

# Set the fraction of data to use for validation
validation_fraction = 0.2

# Get the list of image file paths and corresponding label file paths
image_paths = []
label_paths = []
for filename in os.listdir(os.path.join(input_dir, "images")):
    if filename.endswith(".jpg"):
        image_path = os.path.join(input_dir, "images", filename)
        label_path = os.path.join(input_dir, "labels", filename[:-4] + ".txt")
        if os.path.exists(label_path):
            image_paths.append(image_path)
            label_paths.append(label_path)

# Split the data into training and validation sets
train_image_paths, validation_image_paths, train_label_paths, validation_label_paths = \
    train_test_split(image_paths, label_paths, test_size=validation_fraction)

# Copy the training data to the output directory
train_output_dir = os.path.join(output_dir, "train_1")
os.makedirs(train_output_dir, exist_ok=True)
os.makedirs(os.path.join(train_output_dir, "images"), exist_ok=True)
os.makedirs(os.path.join(train_output_dir, "labels"), exist_ok=True)
for i, (image_path, label_path) in enumerate(zip(train_image_paths, train_label_paths)):
    shutil.copy(image_path, os.path.join(train_output_dir, "images", f"{i}.jpg"))
    shutil.copy(label_path, os.path.join(train_output_dir, "labels", f"{i}.txt"))

# Copy the validation data to the output directory
validation_output_dir = os.path.join(output_dir, "validation")
os.makedirs(validation_output_dir, exist_ok=True)
os.makedirs(os.path.join(validation_output_dir, "images"), exist_ok=True)
os.makedirs(os.path.join(validation_output_dir, "labels"), exist_ok=True)
for i, (image_path, label_path) in enumerate(zip(validation_image_paths, validation_label_paths)):
    shutil.copy(image_path, os.path.join(validation_output_dir, "images", f"{i}.jpg"))
    shutil.copy(label_path, os.path.join(validation_output_dir, "labels", f"{i}.txt"))
```

```python
import os

# Define the current names of the two classes
class1_current_name = "0"
class2_current_name = "1"

# Define the new names of the two classes
class1_new_name = "1"
class2_new_name = "0"

# Define the path to the labels folder
labels_folder = "D:/research/rdd2022/RDD2022_all_countries/test_output/labels - Copy"

# Loop through all the label files in the labels folder
for label_file in os.listdir(labels_folder):
    if label_file.endswith(".txt"):
        # Read the contents of the label file
        with open(os.path.join(labels_folder, label_file), "r") as f:
            label_contents = f.read()
        
        # Replace the current names of the two classes with their new names in the label contents
        label_contents = label_contents.replace(class1_current_name, "___TEMP___")
        label_contents = label_contents.replace(class2_current_name, class1_new_name)
        label_contents = label_contents.replace("___TEMP___", class2_new_name)
        
        # Write the modified label contents back to the label file
        with open(os.path.join(labels_folder, label_file), "w") as f:
            f.write(label_contents)
```

```python
import os

# Define a mapping of old class IDs to new class IDs
class_id_mapping = {
    "0": "1",  # old class ID 0 -> new class ID 2
    "1": "0",  # old class ID 1 -> new class ID 3

}

# Define the path to the labels folder
labels_folder = "D:/research/rdd2022/RDD2022_all_countries/test_output/labels - Copy"

# Loop through all the label files in the labels folder
for label_file_name in os.listdir(labels_folder):
    # Check if the file is a YOLO format label file
    if label_file_name.endswith(".txt"):
        label_file_path = os.path.join(labels_folder, label_file_name)
        with open(label_file_path, "r") as label_file:
            label_contents = label_file.readlines()

        # Modify the class IDs in the label contents
        for i, line in enumerate(label_contents):
            line_parts = line.split()
            old_class_id = line_parts[0]
            if old_class_id in class_id_mapping:
                new_class_id = class_id_mapping[old_class_id]
                line_parts[0] = new_class_id
            label_contents[i] = " ".join(line_parts) + "\n"

        # Write the modified label contents back to the label file
        with open(label_file_path, "w") as label_file:
            label_file.writelines(label_contents)

```
